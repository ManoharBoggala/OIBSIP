```python
df['title'] = df['title'].str.lower()
df.drop_duplicates(inplace=True)
```

```python
df.title
```

```python
outliers=df[(df['views']<0) | (df['likes']>10000)]
print(outliers)
```

```python
df = df[(df['views'] >= 0) & (df['likes'] <= 10000)]
print(df)
```

```python
# Check consistency of related columns
consistency_check = df.groupby('likes')['video_id'].nunique()
print(consistency_check)

# Cross-verification with an external source (if available)
reference_data = pd.read_csv('USvideos.csv')
df = df[df['likes'].isin(reference_data['video_id'])]
```

```python
# Count the number of duplicate rows
num_duplicates = df.duplicated().sum()
print(f'Number of duplicate rows: {num_duplicates}')
```

```python
# Remove duplicate rows, keeping the first occurrence
df_cleaned = df.drop_duplicates()

# Alternatively, remove duplicates based on specific columns
df_cleaned = df.drop_duplicates(subset=['Column1', 'Column2'])

# Preview the cleaned data
df_cleaned.head()
```

```python
# Verify that duplicates have been removed
print(f'Number of rows after duplicate removal: {df.shape[0]}')
```

```python
df_cleaned.to_csv('USvideos.csv', index=False)
```
